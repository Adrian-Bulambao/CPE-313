#### Procedures
Load the necessary libraries

```python
import nltk
import pandas as pd
```

```python
data = pd.read_csv('songdata.csv')
data.head()
```

Find the lyrics by the artist The Beatles

```python
beatles = data[data.artist=='The Beatles']
beatles.head()
```

Check the number of Beatles songs in the data set


```python
data.artist.value_counts().loc[['The Beatles']]
```

Take a look at the first song's lyrics

Remove characters from the lyrics

```python
beatles
```

```python
import re

newline = lambda x: re.sub('\n', ' ', x) # remove \n
alphanumeric = lambda x: re.sub('\w*\d\w*', ' ', x) # remove alphanumeric words

corpus = beatles.text.map(newline).map(alphanumeric)
corpus.head()
```

Create a document term matrix using Count Vectorizer with the stop words turned on to English

```python
from sklearn.feature_extraction.text import CountVectorizer
          
cv = CountVectorizer(stop_words="english")
X = cv.fit_transform(corpus).toarray()

dt = pd.DataFrame(X, columns=cv.get_feature_names_out()).set_index(beatles.song)
dt.head()
```

Set display options so that the lyrics aren't cut off

```python
pd.set_option('display.max_colwidth', None)
```

```python
beatles[beatles.song=='Imagine'].text
```

Imagine lyrics in Count Vectorizer form


```python
imagine = list(dt.loc['Imagine'])
imagine[:20]
```

Define the cosine similarity calculation

```python
from numpy import dot
from numpy.linalg import norm

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))
```

Calculate all similarities and sort by the most similar

```python
similarity = [cosine(imagine, song) for song in X]
sorted(list(zip(similarity, beatles.song)), reverse=True)[1:6]
```

Create the TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
          
cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
          
cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

Calculate all similarities and sort by the most similar using the TF-IDF data

```python
similarity_tfidf = [cosine(imagine, song) for song in X_tfidf]
sorted(list(zip(similarity_tfidf, beatles.song)), reverse=True)[1:6]
```

#### **Supplementary Activity**

* Use a dataset that consists of text document

* Compare the similar document using Count Vector and TF-IDF Vectorizer

The dataset used for the supplementary is a poem dataset from kaggle, taken from this link: https://www.kaggle.com/datasets/mexwell/poem-dataset , each poem has a class, and a total of 6 classes.

```python
import pandas as pd
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords
from nltk.tag import pos_tag
import re

# loading the dataset
poem = pd.read_csv("final_df_emotions(remove-bias).csv")
poem['title'] = ["Poem " + str(i+1) for i in range(len(poem))]
poem = poem [['label', 'title', 'poem content']]
poem.head()
```

* Tokenize (words)

```python
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
stop_words = set(stopwords.words('english'))
tokenizer = TreebankWordTokenizer()
```

```python
poem['tokens_basic'] = poem['poem content'].map(lambda x: tokenizer.tokenize(x))
poem.head(1)
```

* Tokenize (words with remove punctuation) / Tokenize (words with remove special characters and numbers)

```python
clean_text = lambda x: re.sub(r'[^A-Za-z\s]', ' ', x)
poem['tokens_clean'] = poem['poem content'].map(clean_text).map(lambda x: x.split())
poem.head(1)
```

* Tokenize (words with remove punctuation and all lowercase)

```python
poem['tokens_lower'] = poem['poem content'].map(clean_text).map(lambda x: x.lower().split())
poem.head(1)
```

* Tokenize (words with remove punctuation, all lowercase and remove stop words)


```python
poem['tokens_no_stop'] = poem['poem content'].map(clean_text).map(lambda x: x.lower().split()).map(lambda words: [w for w in words if w not in stop_words])
poem.head(1)
```

* Tokenize (words with remove punctuation, all lowercase and  remove stop words, with speech_tagging)

```python
pos_lambda = lambda x: nltk.pos_tag(x)
poem['tokens_pos'] = poem['tokens_no_stop'].map(pos_lambda)
poem.head(1)
```

* Compare the similar document using Count Vector and TF-IDF Vectorizer

```python
# for this part I will take a class of poems and use those classes for the count vector and tf-idf vectorizer
# the sadness class is the one with the most amount of poems in the dataset, so I will be using that  class

sad_poems = poem[poem.label == "sadness" ]

sad_poems['label']
```

```python
sad_poems['cleaned_content'] = sad_poems['tokens_no_stop'].apply(lambda tokens: " ".join(tokens))
corpus1 = sad_poems['cleaned_content']
corpus1.head()
```

```python

count_vect = CountVectorizer()

X_count = count_vect.fit_transform(sad_poems['cleaned_content'])

X_count_df = pd.DataFrame(X_count.toarray(), columns=count_vect.get_feature_names_out(), index=sad_poems['title'])
X_count_df.head()
```

```python
Poem1 = list(X_count_df.loc['Poem 1'])
Poem1[:20]
```

```python
from numpy import dot
from numpy.linalg import norm

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))
```

```python
similarity = [cosine(Poem1, song) for song in X]
sorted(list(zip(similarity, beatles.song)), reverse=True)[1:6]
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(sad_poems['cleaned_content']).toarray()
dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out())
dt_tfidf.index = ["Poem " + str(i+1) for i in range(len(sad_poems))]
dt_tfidf.head()
```

```python
from numpy import dot
from numpy.linalg import norm

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))
```

```python
Poem1_vec = X_tfidf[0]
similarity_tfidf = [cosine(Poem1_vec, vec) for vec in X_tfidf]
sorted(list(zip(similarity_tfidf, dt_tfidf.index)), reverse=True)[1:6]
```

From the result above, it can be seen that Poem 1 is most similar to Poem 58 with a similarity score of 22.97%, indicating that they share some common words. Other poems with lower similarity scores suggest that even though they belong to the same class, their vocabulary and expressions differ more than the poem 58.

* Use the cosine similarity to measure the similarity of text

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
corpus = sad_poems['cleaned_content'].tolist()
vectorizer = CountVectorizer()
X_count = vectorizer.fit_transform(corpus)
cos_sim_matrix = cosine_similarity(X_count)
cos_sim_df = pd.DataFrame(cos_sim_matrix, index=sad_poems['title'], columns=sad_poems['title'])
cos_sim_df.head()
```
