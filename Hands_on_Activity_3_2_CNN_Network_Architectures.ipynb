```python
def save_webcam_image(img_path):

    try:
        camera_port = 0
        ramp_frames = 10

        camera = cv2.VideoCapture(camera_port)

        for i in range(ramp_frames):
            retval, im_camera = camera.read()

        retval, im_camera = camera.read()

        im = cv2.resize(im_camera, (224, 224)).astype(np.float32)
        cv2.imwrite(img_path, im)
        del (camera)
        return True
    except ValueError as e:
        print("Image Capture Failed")
    return False
```

Use VGG-16 pre-trained model

```python
from keras.applications import vgg16


vgg16_model = vgg16.VGG16(weights='imagenet')
vgg16_model.summary()
```

Create a function to load model, image, preprocess input and targets

```python
img_path = "/content/drive/MyDrive/Elective 3/lynx.jpg"


def predict_image(model, img_path, preprocess_input_fn, decode_predictions_fn, target_size=(224, 224)):

    img = image.load_img(img_path, target_size=target_size)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input_fn(x)

    preds = model.predict(x)
    predictions_df = pd.DataFrame(decode_predictions_fn(preds, top=10)[0])
    predictions_df.columns = ["Predicted Class", "Name", "Probability"]
    return predictions_df
```

Predict the result

```python
predict_image(vgg16_model, img_path, vgg16.preprocess_input, vgg16.decode_predictions)
```

Interpret the result of prediction using VGG16 model

#The VGG16 was has predicted the picture and was able to get a 99% probability which is correct, I think that this has to to with it having a lot of parameters in the laters, and those helped it to guess better.

Use Resnet50 pretrained model

```python
from keras.applications import resnet50

resnet50_model = resnet50.ResNet50(weights='imagenet')

resnet50_model.summary()
```

Predict the result

```python
predict_image(resnet50_model, img_path, resnet50.preprocess_input, resnet50.decode_predictions)
```

Interpret the result using RESNET50 pretrained model

#The RESNET50 was has predicted the picture and was able to get a 97% probability which is correct, it has a lower probability score as it has lower parameters and smaller in size but still being able to get a more the 95% which makes it better than the VGG16.

## Using Inception

```python
from keras.layers import Input
from tensorflow.keras.applications import inception_v3


input_tensor = Input(shape=(224, 224, 3))

InceptionV3_model = inception_v3.InceptionV3(input_tensor=input_tensor, weights='imagenet', include_top=True)
InceptionV3_model.summary()
```

```python
predict_image(InceptionV3_model, img_path, inception_v3.preprocess_input, inception_v3.decode_predictions)
```

#The Inception was has predicted the picture and was able to get a 99% probability which is correct, It is a bit better than the VGG16 and smaller in size and parameters needed. It is better than the previous two.

## Using Xception

```python
from tensorflow.keras.applications import xception

input_tensor = Input(shape=(224, 224, 3))


Xception_model = xception.Xception(weights='imagenet', input_tensor= input_tensor)

Xception_model.summary()
```

```python
predict_image(Xception_model, img_path, xception.preprocess_input, xception.decode_predictions)
```

#The Xception was has predicted the picture and was able to get a 99% probability which is correct, it has lower parameters and size but was still able to guess correctly with a high probability score.

## Using Mobilenet

```python
from tensorflow.keras.applications import mobilenet_v2
input_tensor = Input(shape=(224, 224, 3))


MobileNetV2_model = mobilenet_v2.MobileNetV2(weights='imagenet', input_tensor= input_tensor)

MobileNetV2_model.summary()
```

```python
predict_image(MobileNetV2_model, img_path, mobilenet_v2.preprocess_input, mobilenet_v2.decode_predictions)
```

#The MobileNetV2 was has predicted the picture and was able to get an 86% probability, while it is correct, It has a lower probability than all the previous models, as it is much smaller and fewer parameters.
